```python
# 4.Write a pandas program to create a dataframe(6 records) which has the fields id,name,age,degree,cgpa

import numpy as np
import pandas as pd
student={
    "id":[100,101,102,103,104,105],
    "name":["anu","savitha","malavika","amala","sam","karthik"],
    "age":[20,23,18,19,np.nan,34],
    "degree":["BCA","BA","B.TEC",np.nan,"BSC","B.COM"],
    "cgpa":[7.00,8,4,9,np.nan,np.nan]
}

df1=pd.DataFrame(student)
print( " dataframe with 6 records \n\n ",df1)

#   4.1)insert a new column ‘class’ into the dataframe and populate values

df1.insert(2, "class", [1, 3, 2, 4,7,8], True)
print(" \n insert a new column class ","\n *************************\n",df1)
print("A ********************************************* A")

#   4.2)Find out total no.of columns in the dataframe

print("Total no.of columns in the dataframe : ",len(df1.columns))
print("B ********************************************* B")

#   4.3)select top 3 rows ,last 4 rows

print("top 3 rows  : \n",df1.head(3))
print("\n last 4 rows :  \n ",df1.tail(3))
print("C ********************************************* C")

#   4.4)delete all rows which have null values

print("Before removing : \n" ,df1)
print("                                       \n")
df3=df1.dropna()
print("After removing:\n",df3)
print("D ********************************************* D")

#   4.5)drop ‘id’ column

newdf = df1.drop("id", axis='columns')

print(df1)
print("              ")
print( "Remove id : \n ",newdf)
print("                                 ")
print("E ********************************************* E")

#   4.6)select name,cgpa from 2 to 5 records

print(df1.loc[[2,3,4,5],["name","cgpa"]])
print("F ********************************************* F")

#   4.7)fill null values in the age column with mean value

print("Before filling null value \n ",df1)
meanVal = df1['age'].mean()
print("                   ")
print("MEAN VALUE",meanVal)
df1['age'].fillna(value=meanVal,inplace=True)
print("             ")
print(df1)
print("G ********************************************* G")
```
