```python
import pandas as pd
from matplotlib import pyplot as plt
df=pd.read_csv("/home/student/Documents/DSML ABIN/Sample Files/iris.csv")
x=df['sepal_length']
plt.hist(x,bins=20,color="blue")
plt.title("Sepal length")
plt.xlabel("Sepal length")
plt.ylabel("Count")
```

```python
plt.pie(df['species'].value_counts(),labels=['setosa','versicolor','virginica'])
```

```python
df.plot(kind='scatter',x='sepal_length',y='sepal_width')
```

```python
ax1=df.plot(kind='scatter',x='petal_length', y='petal_width',label='petal_length vs petal_width',color='Red')
ax2=df.plot(kind='scatter',x='petal_length', y='sepal_length',label='petal_length vs sepal_length',color='Blue',ax=ax1)
ax3=df.plot(kind='scatter',x='petal_length', y='sepal_width',label='petal_length vs sepal_width', color='Green',ax=ax2)
print(ax1 == ax2 == ax3)
```

```python
from pandas.plotting import scatter_matrix
scatter_matrix(df,alpha=0.5,diagonal='kde')
scatter_matrix(df,alpha=0.5,diagonal='hist',color='Red')
```

```python
plt.scatter('petal_length','petal_width',s='sepal_length',alpha=0.5,data=df)
plt.title("Bubble plot with Matplotlib",size=18)
plt.xlabel("petal_length",size=16)
plt.ylabel("petal_width",size=16)
```
